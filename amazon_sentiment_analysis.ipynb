```python
#This is used to perform sentiment analysis on texts using the VADER sentiment analysis tool.
from nltk.sentiment.vader import SentimentIntensityAnalyzer
s = SentimentIntensityAnalyzer()
```

```python
#importing the data file 
amazon_file=pd.read_csv("amazon.tsv", sep='\t')
amazon_file
```

```python
#analysing the row 1 of the comments 
s.polarity_scores(amazon_file.loc[0]['review'])
```

```python
#adding a column name polarity score to review all comments of the data 
amazon_file['polarity_score']=amazon_file['review'].apply(lambda review:s.polarity_scores(review))
amazon_file
```

```python
# making another column from ploarity score name polarity_score_compound to get compound value from polarity_score column as if polarity scotre is positive comment is positive and if it is negative the comment is negative
amazon_file['polarity_score_compound']=amazon_file['polarity_score'].apply(lambda x:x ['compound'])
amazon_file
```

```python
#getting the remarks of comment as positive or negavtive in senitiment column 
amazon_file['sentiment']= amazon_file['polarity_score_compound'].apply(lambda x:'pos' if x>= 0 else 'neg')
amazon_file
```

```python
#checking the perdicited column(sentiment) from the given one(label) and checking the accuracy
from sklearn.metrics import accuracy_score , classification_report, confusion_matrix
accuracy_score(amazon_file['label'], amazon_file['sentiment'])
#are predicted column as predicted the sentiment of review with approx 70 percent accuracy
```

```python
confusion_matrix(amazon_file['label'],amazon_file['sentiment'])
#A confusion matrix is a table that is often used to evaluate the performance of a classification model. In this case, the confusion matrix represents a binary classification problem where there are two classes: class 0 and class 1.

#1363 represents the number of true negatives (TN). It indicates the number of instances that were actually in class 0 and correctly classified as class 0.

#1199 represents the number of false positives (FP). It indicates the number of instances that were actually in class 0 but incorrectly classified as class 1.

#193 represents the number of false negatives (FN). It indicates the number of instances that were actually in class 1 but incorrectly classified as class 0.

#2009 represents the number of true positives (TP). It indicates the number of instances that were actually in class 1 and correctly classified as class 1.
```
